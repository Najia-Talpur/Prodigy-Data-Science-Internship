```python
import pandas as pd
```

```python
data = pd.read_csv("bank-full.csv", delimiter=';', quotechar='"')
data.head()
```

```python
features = data.drop("y", axis="columns")
target = data[['y']]
```

```python
features.head()
```

```python
target.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le_job = LabelEncoder()
le_marital = LabelEncoder()
le_education = LabelEncoder()
le_default = LabelEncoder()
le_housing = LabelEncoder()
le_loan = LabelEncoder()
le_contact = LabelEncoder()
le_month = LabelEncoder()
le_poutcome = LabelEncoder()
le_y = LabelEncoder()
```

```python
features['job_n'] = le_job.fit_transform(features['job'])
features['marital_n'] = le_marital.fit_transform(features['marital'])
features['education_n'] = le_education.fit_transform(features['education'])
features['default_n'] = le_default.fit_transform(features['default'])
features['housing_n'] = le_housing.fit_transform(features['housing'])
features['loan_n'] = le_loan.fit_transform(features['loan'])
features['contact_n'] = le_contact.fit_transform(features['contact'])
features['month_n'] = le_month.fit_transform(features['month'])
features['poutcome_n'] = le_poutcome.fit_transform(features['poutcome'])
target['y_n'] = le_y.fit_transform(target['y'])
```

```python
features = features.drop(['job','marital','education','default','housing','loan','contact','month','poutcome'], axis='columns')
features.head()
```

```python
target = target.drop('y', axis='columns')
target.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2)
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(X_test, y_test)
```

```python
model.score(X_test, y_test)
```
