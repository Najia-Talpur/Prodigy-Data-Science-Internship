```python
# Generating random colors for each histogram
num_countries = len(countries)
colors = np.random.rand(num_countries, 3)  # Generating random RGB colors

# Plotting histograms for each country's population over the years in 3 graphs per row
num_plots = len(countries)
plots_per_row = 3
num_rows = -(-num_plots // plots_per_row)  # Ceiling division to determine number of rows

fig, axs = plt.subplots(num_rows, plots_per_row, figsize=(15, 5 * num_rows))  # Adjust figure size here
fig.subplots_adjust(hspace=0.5)  # Adjust horizontal space between subplots

for i, country in enumerate(countries):
    row = i // plots_per_row
    col = i % plots_per_row
    ax = axs[row, col] if num_rows > 1 else axs[col]  # Handling single row case
    
    ax.hist(years_data.iloc[i].dropna(), bins=20, color=colors[i])  # Assigning random color
    ax.set_xlabel('Year')
    ax.set_ylabel('Population')
    ax.set_title(f'Population Distribution for {country}')

# Hide empty subplots if any
for i in range(num_plots, num_rows * plots_per_row):
    row = i // plots_per_row
    col = i % plots_per_row
    axs[row, col].axis('off')

plt.show()
```
